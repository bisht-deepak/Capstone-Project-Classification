```python
import numpy as np
import pandas as pd

pd.set_option('display.max_columns', 500)

import matplotlib.pyplot as plt
import seaborn as sns

#    #1f77b4
```

```python
df= pd.read_csv("/kaggle/input/credit-card-defaulter-prediction/Credit Card Defaulter Prediction.csv").drop(columns= ["ID"])
df.head()
```

```python
#rewriting the column names
df.columns= df.columns.str.lower().str.strip()
df.columns
```

```python
df.isnull().sum()
```

```python
df.describe()
```

```python
df.describe(include= "object")
```

```python
#eda

#univariate analysis


numerical_cols= list(df.select_dtypes(np.number).columns)

num_categorical_cols= ["pay_0", "pay_2", "pay_3", "pay_4", "pay_5", "pay_6"]

numerical_cols = [i for i in numerical_cols if i not in num_categorical_cols]
```

Numerical Features- "pay_0", "pay_2", "pay_3", "pay_4", "pay_5", "pay_6", look more like categories

```python
fig, axes= plt.subplots(4, 4, figsize= (30, 20))
axes= axes.flatten()

# plt.subplots_adjust(wspace= 0.5, hspace= 0.5)

for i, col in enumerate(numerical_cols):
    sns.histplot(data= df, x= col, kde= True, ax= axes[i])
    axes[i].set_title(col, fontsize= 15)
    
for i in range(len(numerical_cols), 16):
    fig.delaxes(axes[i])

    
plt.suptitle("Distribution of Numerical Features", fontsize= 20)

plt.tight_layout()
plt.show()
```

```python
from scipy.stats import probplot

fig, axes= plt.subplots(4, 4, figsize= (30, 25))
axes= axes.flatten()

for i, col in enumerate(numerical_cols):
    probplot(x= df[col], dist= "norm", plot= axes[i])
#     probplot[0].set_markerfacecolor('#1f77b4')
    axes[i].set_title(col, fontsize= 15)
    

for i in range(len(numerical_cols), 16):
    fig.delaxes(axes[i])
    
plt.suptitle("QQ plot for numerical cols", fontsize= 20)

plt.tight_layout()
plt.show()
```

```python
#visualizing outliers
fig, axes= plt.subplots(4, 4, figsize= (30, 20))
axes= axes.flatten()

for i, col in enumerate(numerical_cols):
    sns.boxplot(data= df, x= col, ax= axes[i])
    axes[i].set_title(col, fontsize= 15)

for i in range(len(numerical_cols), 16): 
    fig.delaxes(axes[i])

plt.suptitle("Outliers Visualization for Numerical Features", fontsize= 20)
plt.tight_layout()
plt.show()
```

```python
categorical_cols= list(df.select_dtypes("object").columns)
categorical_cols.extend(num_categorical_cols)
categorical_cols
```

```python
fig, axes= plt.subplots(3, 4, figsize= (30, 20))
axes= axes.flatten()

for i, col in enumerate(categorical_cols):
    sns.countplot(data= df, x= col, color= "#1f77b4", ax= axes[i])
    axes[i].set_title(col)

for i in range(len(categorical_cols), 12):
    fig.delaxes(axes[i])

plt.suptitle("Distribution of Categorical Features", fontsize= 20)
    
plt.tight_layout()
plt.show()
```

```python
pie_df= df.groupby(["default"]).size().reset_index().rename(columns= {0: "size"})
pie_df
```

```python
#bivariate analysis

#violin plots

fig, axes= plt.subplots(4, 4, figsize= (30, 20))
axes= axes.flatten()

for i, col in enumerate(numerical_cols):
    sns.violinplot(
        data= df, 
        x= col, y= "default", scale= "count",
        ax= axes[i])
    axes[i].set_title(col)
#     axes[i].set_ylabel("default", fontsize= 10)
    
for i in range(len(numerical_cols), 16):
    fig.delaxes(axes[i])
    
plt.suptitle("Violin Plot for Default v/s", fontsize= 20)

plt.tight_layout()
plt.show()
```

```python
#kde plot

fig, axes= plt.subplots(4, 4, figsize= (30, 20))
axes= axes.flatten()

for i, col in enumerate(numerical_cols):
    sns.kdeplot(data= df, x= col, hue= "default",
                fill= True, ax= axes[i])
    axes[i].set_title(col)
    
for i in range(len(numerical_cols), 16):
    fig.delaxes(axes[i])
    
plt.suptitle("KDE Plot for Default v/s", fontsize= 20)

plt.tight_layout()
plt.show()
```
